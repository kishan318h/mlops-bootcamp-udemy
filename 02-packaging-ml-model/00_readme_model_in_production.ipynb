## Model Development

We create the model in a jupyter notebook. It cannot be used directly in a production environment because
1. Server may not have a GUI to debug the code
2. There are code redundancies
3. The development was not done in virtual enviroment.
    - We need to ensure that the version of package used for development is present in the prod
    - Otherwise, it might lead to errors

## Step 1: Create a virtual environment

use `venv` to create a virtual environment

Create requirements file. It is a text file which contains the list of all libraries used in the development of model

to create requirements file: `pip freeze > requirements.txt`
to use the file: `pip install -r requirements.txt`

## Step 2: Modularize the code

Modular programming is an approach where 
    - the code is divided into separate files
    - each file has everything necessary to execute a defined piece of logic
    - it returns expected output when imported by other files as input for them

A module is a python file that can hold classes, functions, variables, etc.

Package in python:

- one or more modules can be interlinked with each other
- a directory with subdirectory can be called as package if it contains `__init__.py` file

### Structure of Package A

Package A
- __init__.py
- func1.py
- func2.py
- Sub-package 1
    - __init__.py
    - func3.py
    - func4.py
- Sub-package 2
    - __init__.py
    - func5.py


### How to import from above package

- from packageA import func1
- from packageA.subpackage1 import func3, func4
- from packageA.subpackage2 import func5

### Module paths

A module can only be imported when the module is present in a folder where the program is searching for it. To check all the paths where the code searches for a module. Use 'sys' library

```python
import sys
sys.path
```

The output is showing a list of path where the code searches for any module

We can add our own custom path using `sys.path.append`

Syntax: `sys.path.append('/opt/anaconda3/envs/ML-DL-Learn/lib/')`
