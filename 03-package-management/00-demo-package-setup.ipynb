## Step 1

- Create virtual environment using the following command
`conda create -n env_name python=3.12`
- Activate virtual environment: `conda activate env_name`
- install setuptools: `pip install setuptools twine`

## Step 2

Create the folder structure:

- hello-world
    - |_ setup.py
    - |_ hello_world #folder
        - |_ `__init__.py`
        - |_ main.py

`setup.py` and `__init__.py` are mandatory files

## Step 3:
Code to generate a whl file for the package using setuptools.

~~~
from setuptools import setup, find_packages

setup(
    name="hello_world_kk318",
    version="0.0.1",
    author="ichiraku_k",
    author_email="kishan.gitrepo@gmail.com",
    description="A hello-world example package",
    packages=find_packages(),
    classifiers=[
        "Programming Language :: Python :: 3",
        "License :: OSI Approved :: MIT License",
        "Operating System :: OS Independent",
    ],
)
~~~

- Write functions in the `main.py`
- in `__init__.py` write `from .main import function_name`
    - this import statement helps in importing the function from the package
- Run the 'setup.py' file in terminal using `python setup.py sdist`
    - creates a source distribution file - setup file
    - the setup file will be create in a folder named dist. The file will have an extension `.tar.gz`



## Step 4:
Install the setup file: `pip install dist/filename.tar.gz`
import the created module and use in python

## Step 5:
Create an account in `test.pypi.org` if not created already. Setup by generating token. Upload the library to PiPy test server: `twine upload --repository-url https://test.pypi.org/legacy/ folder/filename.tar.gz`
